```python
%pip install speechrecognition
%pip install pyaudio
import speech_recognition as sr
from gtts import gTTS
import os

# Initialize recognizer instance
r = sr.Recognizer()


def record_audio():
    # Define microphone as source for audio input
    with sr.Microphone() as source:
        print("Speak something:")
        # Adjust for ambient noise levels
        r.adjust_for_ambient_noise(source)
        # Listen for user input
        audio = r.listen(source)
    try:
        # Use Google speech recognition API to convert audio to text
        text = r.recognize_google(audio)
        print("You said: " + text)
        return text
    except sr.UnknownValueError:
        print("Could not understand audio")
        return ""
    except sr.RequestError as e:
        print(
            "Could not request results from Google Speech Recognition service; {0}".format(e))
        return ""


def text_to_speech(text):
    # Use gTTS API to convert text to speech
    tts = gTTS(text=text, lang='en')
    # Save audio file
    tts.save("audio.mp3")
    # Play audio file
    os.system("mpg321 audio.mp3")


# Assistant logic
while True:
    # Record audio from user
    text = record_audio()
    if text:
        # Determine appropriate response
        if "hello" in text:
            response = "Hello there!"
        elif "what is the time" in text:
            # Use datetime library to retrieve current time
            import datetime
            response = "The time is " + datetime.datetime.now().strftime("%I:%M %p")
        elif "what is the weather like" in text:
            # Call weather API to retrieve weather information
            import requests
            api_key = "your-api-key-here"
            base_url = "http://api.openweathermap.org/data/2.5/weather?"
            city_name = "New York"
            complete_url = base_url + "appid=" + api_key + "&q=" + city_name
            response_data = requests.get(complete_url).json()
            temperature = response_data["main"]["temp"]
            response = "The temperature is " + \
                str(round(temperature - 273.15, 1)) + " degrees Celsius"
        else:
            response = "I'm sorry, I don't understand what you're asking for."
        # Convert response to speech and play audio
        text_to_speech(response)
```
